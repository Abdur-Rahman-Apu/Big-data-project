```python
# find first name
def find_fname(name):
  name=name.split();
  size=len(name);
  if (size==2):
      return name[0]
  elif (size==3):
      return name[0]+' '+name[1]
  elif (size==4):
    return name[0]+' '+name[1]+' '+name[2]


split_first_name=F.udf(find_fname,StringType());
first_name=first_name.withColumn('first_name',split_first_name('Employee Name'))
first_name.show()
```

```python
#find last name
def find_lname(name):
  name=name.split();
  size=len(name);
  return name[size-1]


split_last_name=F.udf(find_lname,StringType());
name=first_name.withColumn('last_name',split_last_name('Employee Name'))
name.show()
```
